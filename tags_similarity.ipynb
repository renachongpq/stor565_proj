```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
genome_scores = pd.read_csv("./ml-25m/genome-scores.csv")
```

```python
table = pd.pivot_table(genome_scores, values="relevance", index="movieId", columns="tagId")
```

```python
table
```

```python
similarity_arr = cosine_similarity(table)
```

```python
movie_id = pd.Series(table.index.values, name="movieId")
similarity = pd.concat([movie_id, pd.DataFrame(similarity_arr)], axis=1)
similarity.set_index("movieId", inplace=True)
```

```python
similarity.rename(mapper=lambda x: movie_id[x], axis=1, inplace=True)
```
